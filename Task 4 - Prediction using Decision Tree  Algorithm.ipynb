## Decision Tree Classification
For the given 'Iris' dataset, create the Decision Tree classifier and visualize it graphically.

## Exploratory Data Analysis
### Python Libraries being used:-
#### A.Data analysis

```python
import pandas as pd
import numpy as np
```

#### B. Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#### C. Model Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
from sklearn import metrics
```

### Data Exraction or Loading Data
- Importing dataset.
- 1 files:-
    - iris data -- our training file.

```python
from sklearn import datasets
```

```python
iris_data = datasets.load_iris()
iris = pd.DataFrame(iris_data['data'],columns=['sepal_lengthCm', 'sepal_widthCm', 'petal_lengthCm', 'petal_widthCm'])
```

```python
iris['Species']=iris_data['target']
iris['Species']=iris['Species'].apply(lambda x: iris_data['target_names'][x])
iris.head()
```

## DATA VISUALIZATION

```python
sns.pairplot(iris, hue = 'Species', palette="Set1")
plt.show()
```

```python
d = ["sepal_lengthCm", "sepal_widthCm", "petal_lengthCm", "petal_widthCm"]
for d in d:
    sns.catplot(data=iris, x=d, y="Species")
    plt.show()
```

Observed that iris-setosa species is clearly separable from the other species.

### Correlation plot:-

```python
plt.figure(figsize=(10,10))
sns.heatmap(iris.corr(),annot=True)
plt.plot()
```

Observed that Petal length is highly related to petal width and Sepal lenth is not related to sepal width.

### Categorical Distribution of Species using BoxPlot

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=iris, x="Species",y="sepal_lengthCm")
plt.subplot(2,2,2)
sns.boxplot(data=iris, x="Species",y="sepal_widthCm")
plt.subplot(2,2,3)
sns.boxplot(data=iris, x="Species",y="petal_lengthCm")
plt.subplot(2,2,4)
sns.boxplot(data=iris, x="Species",y="petal_widthCm")
```

## Data Modelling

### Spliting data into Train and Test sets

```python
train,test = train_test_split(iris,test_size=0.3)
```

### Training

```python
train_X = train[["sepal_lengthCm", "sepal_widthCm", "petal_lengthCm", "petal_widthCm"]]
train_y = train.Species
```

### Testing

```python
test_X=test[["sepal_lengthCm", "sepal_widthCm", "petal_lengthCm", "petal_widthCm"]]
test_y=test.Species
```

```python
dectree = DecisionTreeClassifier(random_state=12)
model = dectree.fit(train_X, train_y)
```

```python
dectree.predict(test_X)
```

### Evaluating Accuracy

```python
dectree.score(test_X, test_y)
```

```python
y_pred = dectree.predict(test_X)
print('Accuracy Score:', metrics.accuracy_score(test_y, y_pred))
```

```python
fig, axes = plt.subplots(nrows = 1,ncols =1,figsize = (11,11), dpi=200)
_ = tree.plot_tree(dectree,feature_names=iris_data.feature_names,class_names=iris_data.target_names,filled=True,fontsize=7)
```
